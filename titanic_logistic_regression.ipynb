# Titanic - Linear Regression

```python
import pandas as pd

train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")

train_data.head()
```

```python
train_data.describe()
```

```python
train_data.isnull().sum()
```

```python
train_data.fillna({"Age": train_data.Age.mean()}, inplace=True)
test_data.fillna({"Age": test_data.Age.mean()}, inplace=True)
```

```python
test_data.fillna({"Fare": test_data.Fare.mean()}, inplace=True)
```

```python
train_data.Sex = train_data.Sex.map({"female": 1, "male": -1}).astype(int)
test_data.Sex = test_data.Sex.map({"female": 1, "male": -1}).astype(int)
```

```python
train_data.Embarked = train_data.Embarked.notna().astype(int)
test_data.Embarked = test_data.Embarked.notna().astype(int)
```

```python
train_data.Cabin = train_data.Cabin.notna().astype(int)
test_data.Cabin = test_data.Cabin.notna().astype(int)
```

```python
train_data = train_data.drop(["Name", "Ticket"], axis=1)
test_data = test_data.drop(["Name", "Ticket"], axis=1)
```

```python
train_data.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']
target = 'Survived'

X_train, X_test, y_train, y_test = train_test_split(train_data[features], train_data[target], test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
test_data.isna().sum()
```

```python
test_predictions = model.predict(test_data[features])

submission = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': test_predictions})
submission.to_csv('submission.csv', index=False)
```
